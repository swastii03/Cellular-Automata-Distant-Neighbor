## Question

You want to send a document to your bank and it is very important it gets there the next day. You therefore send two copies of the document using Hermes (delivery service $A$) and the Royal Mail (delivery service $B$).  The probabilities of those companies delivering on time are:

- 0.90 for service A
- 0.80 for service B
- 0.75 for both services being on time.

Does your strategy improve the chances the document will be delivered on time?

### Answer

```python
Pa= 0.90
Pb=0.80
PaANDb=0.75

PaUNIONb=Pa+Pb-PaANDb
print("Probability of the documents being delivered is {:.2f}". format(PaUNIONb))

if PaUNIONb>Pa:
  print("Yes, sending two copies increases the chances of the delivery of doc on time")
else:
  print("No, sending two copies does not increase the chances of the delivery of doc on time")
```

The probability does seems to increase with both the methods as a union of both the probabilities. The two probabilities could be considered as two different sets and PaANDb is their interection, basically representing the common ground or probability between these two sets. Subtracting them from the addition of Pa and Pb will provide the union or the probability of the document being delivered by A or B on time.
